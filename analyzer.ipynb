```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("input.txt")
```

### Sort the data by keypresses

```python
# sort by presses
df.sort_values(by=['presses'], inplace=True, ascending=False)
df.head()

# find the total presses
total = df['presses'].sum()
print("The total is", total, "presses")

# create a new percent column
# this tells you what percent of presses the key
# makes up
percent = []
for presses in df['presses']:
    percent.append(presses/total * 100)
df['percent'] = percent


df.head()
```

### Displays the number of key presses

```python
sns.set_theme(style='darkgrid', rc={'figure.dpi': 150},              
              font_scale=1)

fig = sns.barplot(x = df["key"] , y = df["presses"])

fig.set_xticklabels(fig.get_xticklabels(),rotation = 90)
fig.set_title("Number Presses Per Key")

fig
```

```python
percent_fig = sns.barplot(x = df["key"] , y = df["percent"], )

percent_fig.set_xticklabels(percent_fig.get_xticklabels(),rotation = 90)
percent_fig.set_title("Number Presses Per Key")

percent_fig
```

```python
# sort the dataframe by letters
# more specifically since all the letters have length of
# 1 we can get all the keys with length 1

# todo add a comparison from other data
letters_df = df[df.key.str.len() == 1]

letters_fig = sns.barplot(x = "key" , y = "presses", data=letters_df)
letters_fig.set_title("Number Presses Per Letter")

letters_fig
```

```python
# similarly we can do it by not letters
other_df = df[df.key.str.len() > 1]
other_fig = sns.barplot(x = "key" , y = "presses", data=other_df)
other_fig.set_xticklabels(other_fig.get_xticklabels(),rotation = 90)
other_fig.set_title("Number Presses Per Other Key")
other_fig

plt.show()
```
